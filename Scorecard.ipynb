```python
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
import pandas as pd
print("imported")
```

```python
url="https://www.sportskeeda.com/live-cricket-score/mumbai-indians-vs-chennai-super-kings-1st-match-19-september-2020"
driver=webdriver.Chrome("C:/Program Files (x86)/chromedriver.exe")
driver.get(url)
```

```python
soup=BeautifulSoup(driver.page_source,'lxml')
#a=soup.select(".one-innings-div")
```

```python
b=soup.select(".innings-batsman a")
name=[]
for i in b:
    name.append(i.text)
df=pd.DataFrame(name)
df.rename(columns={0:'Batsman'},inplace=True)
df
```

```python
bat=soup.select(".innings-table-batting .innings-runs")
runs=[]
for i in bat:
    runs.append(i.text)
runs
df['Runs']=runs
df
```

```python
bowl=soup.select(".innings-bowler a")
bowler=[]
for i in bowl:
    bowler.append(i.text)
bowler
bl=pd.DataFrame(bowler)
bl
```

```python
bl.columns=['Bowler']
bl
```

```python
w=soup.select(".innings-table-bowling .innings-wickets")
m=soup.select(".innings-table-bowling .innings-maiden-overs")
wicks=[]
maiden=[]
for i in w:
    wicks.append(i.text)
for i in m:
    maiden.append(i.text)
bl['Wickets']=wicks
bl['Maiden Overs']=maiden
```
